# Importing Matplotlib

```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

# Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

# Bar Plot

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]

plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

# Histogram

```python
data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# Pie Chart

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

# Customizing Plots

## Adding Titles and Labels

```python
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('Angle [radians]')
plt.ylabel('Sine')
plt.show()
```

## Adding Legends

```python
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Sine')
plt.plot(x, y2, label='Cosine')
plt.legend()
plt.title('Sine and Cosine Waves')
plt.xlabel('Angle [radians]')
plt.ylabel('Value')
plt.show()
```

## Changing Line Styles and Colors

```python
plt.plot(x, y, linestyle='--', color='r', marker='o')
plt.title('Customized Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## Subplots

```python
fig, axs = plt.subplots(2, 2)

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sine')

axs[0, 1].plot(x, y2, 'tab:orange')
axs[0, 1].set_title('Cosine')

axs[1, 0].plot(x, y3, 'tab:green')
axs[1, 0].set_title('Tangent')

axs[1, 1].plot(x, y4, 'tab:red')
axs[1, 1].set_title('Exponential')

for ax in axs.flat:
    ax.set(xlabel='X axis', ylabel='Y axis')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()

plt.show()
```

## Saving Figures

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.savefig('line_plot.png')
```

# Advanced Plotting

## Error Bars

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
yerr = np.random.rand(10) * 0.2

plt.errorbar(x, y, yerr=yerr, fmt='-o')
plt.title('Error Bars')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## 3D Plotting

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Surface Plot')
plt.show()
```

## Interactive Plotting with Widgets

```python
from matplotlib.widgets import Slider

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.25)
line, = plt.plot(x, y)

axcolor = 'lightgoldenrodyellow'
axamp = plt.axes([0.1, 0.1, 0.65, 0.03], facecolor=axcolor)

samp = Slider(axamp, 'Amp', 0.1, 10.0, valinit=1)

def update(val):
    amp = samp.val
    line.set_ydata(amp * np.sin(x))
    fig.canvas.draw_idle()

samp.on_changed(update)
plt.show()
```
